```python
class Node:
    def __init__(self, value):
        self.value = value  # value is a list: [english_word, pronunciation, translations]
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None


    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert_recursive(self.root, value)

    def _insert_recursive(self, current_node, value):
        if value[0] < current_node.value[0]:
            if current_node.left is None:
                current_node.left = Node(value)
            else:
                self._insert_recursive(current_node.left, value)
        elif value[0] > current_node.value[0]:
            if current_node.right is None:
                current_node.right = Node(value)
            else:
                self._insert_recursive(current_node.right, value)
        else:
            print("A palavra já existe na árvore!")

    def find(self, value):
        return self._find_recursive(self.root, value)

    def _find_recursive(self, current_node, value):
        if current_node is None:
            return None
        elif value == current_node.value[0]:
            return current_node.value
        elif value < current_node.value[0]:
            return self._find_recursive(current_node.left, value)
        else:
            return self._find_recursive(current_node.right, value)

    def inorder_traversal(self):
        return self._inorder_traversal_recursive(self.root)

    def _inorder_traversal_recursive(self, current_node):
        result = []
        if current_node is not None:
            result = self._inorder_traversal_recursive(current_node.left)
            result.append(current_node.value)
            result = result + self._inorder_traversal_recursive(current_node.right)
        return result

    def addWord(self, value):
        pass

# Função para carregar dados do CSV e inserir na árvore
import csv

def load_csv_to_bst(filename):
    tree = BinaryTree()
    with open(filename, newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile, delimiter=';')
        next(reader)  # pular o cabeçalho
        for row in reader:
            english_word = row[0]
            pronunciation = row[1]
            translations = row[2]
            tree.insert([english_word, pronunciation, translations])
    return tree

# Função para traduzir uma palavra do inglês para o português usando a árvore
def translate(tree, english_word):
    node = tree.find(english_word)
    if node:
        return node[2]  # Retorna a tradução
    else:
        return "Tradução não encontrada"

# Exemplo de uso
filename = 'translations.csv'  # Substitua pelo caminho do seu arquivo CSV
bst = load_csv_to_bst(filename)

def print_tree(root, val="value", left="left", right="right"):
    def display(root, val=val, left=left, right=right):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if getattr(root, right) is None and getattr(root, left) is None:
            line = '%s' % getattr(root, val)[0]
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Only left child.
        if getattr(root, right) is None:
            lines, n, p, x = display(getattr(root, left))
            s = '%s' % getattr(root, val)[0]
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Only right child.
        if getattr(root, left) is None:
            lines, n, p, x = display(getattr(root, right))
            s = '%s' % getattr(root, val)[0]
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Two children.
        left, n, p, x = display(getattr(root, left))
        right, m, q, y = display(getattr(root, right))
        s = '%s' % getattr(root, val)[0]
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

    lines, *_ = display(root, val, left, right)
    for line in lines:
        print(line)
```
