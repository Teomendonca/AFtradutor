### Classificação

Vamos encontrar uma ou mais regras para identificar as espécies das flores e validar se a regra está funcionando bem.

```python
# primeiro, vamos criar uma coluna para colocar nossas respostas das regras

iris['predicao'] = ''
print(iris.shape)
```

```python
# x-> comprimento_petala
# y-> largura_petala

#criar um grafico com dos dados de treino, identitificando cada resposta

# tambem colocar os valores de teste para comparação

#quanto mais proximo do dado, ele sera esse dado

# usar dados significativos em casos mais complexos para auxiliar no agrupamento

```

```python
setosa = iris[iris['classe'] == 'Iris-setosa']
virginica = iris[iris['classe'] == 'Iris-virginica']
versicolor = iris[iris['classe'] == 'Iris-versicolor']

df = [setosa.max(), virginica.max(), versicolor.max()]
df = pd.DataFrame(df)

dfmin=[setosa.min(), virginica.min(), versicolor.min()]
dfmin = pd.DataFrame(dfmin)

#df
#dfmin
# SETOSA
set_comp_petala_max = setosa['comprimento_petala'].max()
set_comp_petala_min = setosa['comprimento_petala'].min()
set_larg_petala_max = setosa['largura_petala'].max()
set_larg_petala_min = setosa['largura_petala'].min()

set_comp_sepala_max = setosa['comprimento_sepala'].max()
set_comp_sepala_min = setosa['comprimento_sepala'].min()
set_larg_sepala_max = setosa['largura_sepala'].max()
set_larg_sepala_min = setosa['largura_sepala'].min()

#VIRGINICA
iv_comp_petala_max = virginica['comprimento_petala'].max()
iv_comp_petala_min = virginica['comprimento_petala'].min()
iv_larg_petala_max = virginica['largura_petala'].max()
iv_larg_petala_min = virginica['largura_petala'].min()

iV_comp_sepala_max = virginica['comprimento_sepala'].max()
iV_comp_sepala_min = virginica['comprimento_sepala'].min()
iV_larg_sepala_max = virginica['largura_sepala'].max()
iV_larg_sepala_min = virginica['largura_sepala'].min()

#VERSICOLOR
ver_comp_petala_max = versicolor['comprimento_petala'].max()
ver_comp_petala_min = versicolor['comprimento_petala'].min()
ver_larg_petala_max = versicolor['largura_petala'].max()
ver_larg_petala_min = versicolor['largura_petala'].min()

ver_comp_sepala_max = versicolor['comprimento_sepala'].max()
ver_comp_sepala_min = versicolor['comprimento_sepala'].min()
ver_larg_sepala_max = versicolor['largura_sepala'].max()
ver_larg_sepala_min = versicolor['largura_sepala'].min()
```

```python
df
```

```python
# adapte ou adicione novas regras, e execute novamente essa célula
# as petalas são iguais, segundo comentário
iris['predicao'][
    (iris['comprimento_petala'] > set_comp_petala_min) & (iris['comprimento_petala']<set_comp_petala_max) &
    (iris['largura_petala'] > set_larg_petala_min) & (iris['largura_petala']<set_larg_petala_max) &
    (iris['comprimento_sepala'] > set_comp_sepala_min) & (iris['comprimento_sepala']<set_comp_sepala_max) &
    (iris['largura_sepala'] > set_larg_sepala_min) & (iris['largura_sepala']<set_larg_sepala_max)
    ] = 'Iris-setosa'

#virginica
iris['predicao'][
    (iris['comprimento_petala'] >iv_comp_petala_min ) & (iris['comprimento_sepala'] <iv_comp_petala_max)&
    (iris['largura_petala'] > iv_larg_petala_min) & (iris['largura_petala']<iv_larg_petala_max) &
    (iris['comprimento_sepala'] > iV_comp_sepala_min) & (iris['comprimento_sepala'] < iV_comp_sepala_max) &
    (iris['largura_sepala'] > iV_larg_sepala_min) & (iris['largura_sepala']<iV_larg_sepala_max)
    ] = 'Iris-virginica'
#versicolor

iris['predicao'][(iris['comprimento_petala'] > 3) & (iris['comprimento_sepala'] < 5.1)] = 'Iris-versicolor'
iris['predicao'][(iris['largura_petala'] > 1) & (iris['largura_petala']<1.8)] = 'Iris-versicolor'
iris['predicao'][(iris['comprimento_sepala'] > ver_comp_sepala_min) & (iris['comprimento_sepala'] < ver_comp_sepala_max)] = 'Iris-versicolor'
iris['predicao'][(iris['largura_sepala'] > ver_larg_petala_min) & (iris['largura_sepala']<ver_larg_sepala_max)] = 'Iris-versicolor'

```

```python
# as seguintes instruções verificam quantos acertos foram obtidos com as regras

print("Acurácia Geral")
print((iris['classe'] == iris['predicao']).sum() / len(iris))

for classe in list(iris['classe'].unique()):
  print("Acurácia da classe", classe)
  aux = iris[iris['classe'] == classe]
  print((aux['classe'] == aux['predicao']).sum() / len(aux))
```
